```python
import random

def guess_the_number():
    # Generate a random number between 1 and 100
    secret_number = random.randint(1, 100)
    attempts = 0

    print("Welcome to Guess the Number!")
    print("I've selected a number between 1 and 100. Can you guess it?")

    while True:
        guess = int(input("Enter your guess: "))
        attempts += 1

        if guess < secret_number:
            print("Too low! Try again.")
        elif guess > secret_number:
            print("Too high! Try again.")
        else:
            print(f"Congratulations! You've guessed the number {secret_number} in {attempts} attempts!")
            break

        play_again = input("Do you want to play again? (yes/no): ").lower()
        if play_again != 'yes':
            break

guess_the_number()
```

```python
import time

class Room:
    def __init__(self, name, description, items=None):
        self.name = name
        self.description = description
        self.items = items or []
        self.connections = {}

    def add_connection(self, direction, connected_room):
        self.connections[direction] = connected_room

    def __str__(self):
        return self.name


class Player:
    def __init__(self, current_room):
        self.current_room = current_room
        self.inventory = []

    def move(self, direction):
        if direction in self.current_room.connections:
            self.current_room = self.current_room.connections[direction]
            print("Moving to", self.current_room)
            print(self.current_room.description)
        else:
            print("You can't go that way!")


def play_game():
    # Define rooms
    living_room = Room("Living Room", "You are in the living room. It's dark and spooky.")
    kitchen = Room("Kitchen", "You are in the kitchen. You hear strange noises coming from the pantry.")
    pantry = Room("Pantry", "You are in the pantry. It's cold and musty.")
    hallway = Room("Hallway", "You are in the hallway. There are several doors leading to different rooms.")

    # Connect rooms
    living_room.add_connection("east", kitchen)
    kitchen.add_connection("west", living_room)
    kitchen.add_connection("north", pantry)
    pantry.add_connection("south", kitchen)
    hallway.add_connection("south", living_room)
    hallway.add_connection("east", pantry)

    # Create player and start in living room
    player = Player(living_room)

    print("Welcome to the Haunted House! Can you find your way out?")
    print(player.current_room.description)

    # Main game loop
    while True:
        command = input(">").lower().strip()

        if command == "quit":
            print("Thanks for playing!")
            break

        elif command in ["north", "south", "east", "west"]:
            player.move(command)

        else:
            print("Invalid command. Try again.")

play_game()
```

```python
import random

# List of countries and continents
countries = ["Argentina", "Australia", "Brazil", "Canada", "China", "Egypt", "France", "Germany", "India", "Italy", "Japan", "Mexico", "Russia", "Spain", "United States"]
continents = ["Africa", "Antarctica", "Asia", "Europe", "North America", "South America"]

def choose_word():
    # Randomly choose between a country or continent
    word_type = random.choice(["country", "continent"])
    if word_type == "country":
        return random.choice(countries)
    else:
        return random.choice(continents)

def generate_puzzle(word):
    # Shuffle the letters of the word
    word_list = list(word)
    random.shuffle(word_list)
    return ''.join(word_list)

def display_help(word):
    # Display a hint for the word
    if word in countries:
        print("Hint: This is a country.")
    else:
        print("Hint: This is a continent.")

def play_game():
    word = choose_word()
    puzzle = generate_puzzle(word)

    print("Welcome to the Word Puzzle Game!")
    print("Unscramble the letters to guess the word.")
    print("Puzzle:", puzzle)

    while True:
        guess = input("Your guess (or type 'help' for a hint): ").capitalize()

        if guess == "Help":
            display_help(word)
        elif guess == word:
            print("Congratulations! You've guessed the word correctly.")
            break
        else:
            print("Incorrect guess. Try again.")

play_game()
```

```python
import random

class Player:
    def __init__(self, name, health=100, attack=10):
        self.name = name
        self.health = health
        self.attack = attack

    def is_alive(self):
        return self.health > 0

    def take_damage(self, damage):
        self.health -= damage

    def attack_enemy(self, enemy):
        damage = random.randint(1, self.attack)
        enemy.take_damage(damage)
        print(f"{self.name} attacks {enemy.name} for {damage} damage.")

class Enemy:
    def __init__(self, name, health, attack):
        self.name = name
        self.health = health
        self.attack = attack

    def is_alive(self):
        return self.health > 0

    def take_damage(self, damage):
        self.health -= damage

    def attack_player(self, player):
        damage = random.randint(1, self.attack)
        player.take_damage(damage)
        print(f"{self.name} attacks {player.name} for {damage} damage.")

def create_enemy():
    enemies = [
        Enemy("Goblin", 20, 5),
        Enemy("Skeleton", 30, 8),
        Enemy("Orc", 40, 10)
    ]
    return random.choice(enemies)

def explore_dungeon(player):
    print("You enter the dungeon. Adventure awaits!")
    while player.is_alive():
        print("\nYou continue deeper into the dungeon...")
        if random.random() < 0.5:  # 50% chance of encountering an enemy
            enemy = create_enemy()
            print(f"\nOh no! You encounter a {enemy.name}!")
            while enemy.is_alive() and player.is_alive():
                player.attack_enemy(enemy)
                if enemy.is_alive():
                    enemy.attack_player(player)
            if player.is_alive():
                print(f"\nYou defeated the {enemy.name}! You gain some treasure.")
                # Add treasure collection logic here
            else:
                print("\nYou were defeated in battle. Game over.")
                break
        else:
            print("You explore the room but find nothing of interest.")

# Main game loop
def play_game():
    print("Welcome to the Dungeon Adventure!")
    name = input("Enter your character's name: ")
    player = Player(name)
    print(f"Hello, {player.name}! Let's begin your adventure.")
    explore_dungeon(player)

play_game()
```

```python
import random

class PlayfulChatBot:
    def __init__(self):
        self.jokes = [
            "Why don't scientists trust atoms? Because they make up everything!",
            "Why did the scarecrow win an award? Because he was outstanding in his field!",
            "Why did the bicycle fall over? Because it was two-tired!"
        ]
        self.songs = [
            "Twinkle, twinkle, little star, How I wonder what you are!",
            "Row, row, row your boat, Gently down the stream...",
            "Old MacDonald had a farm, E-I-E-I-O!"
        ]
        self.funny_responses = [
            "Haha, that's funny!",
            "LOL!",
            "You cracked me up!",
            "I've got a joke for you too, want to hear it?"
        ]

    def tell_joke(self):
        return random.choice(self.jokes)

    def sing_song(self):
        return random.choice(self.songs)

    def respond_to_query(self, query):
        if "joke" in query.lower():
            return self.tell_joke()
        elif "song" in query.lower():
            return self.sing_song()
        else:
            return random.choice(self.funny_responses)

# Main function to interact with the chatbot
def interact_with_chatbot():
    bot = PlayfulChatBot()
    print("Welcome to the Playful ChatBot!")
    print("You can ask me for jokes, songs, or just chat with me.")
    print("Type 'quit' to exit.")
    while True:
        query = input("You: ")
        if query.lower() == "quit":
            print("Goodbye!")
            break
        response = bot.respond_to_query(query)
        print("ChatBot:", response)

# Start interacting with the chatbot
interact_with_chatbot()
```

##Graph making


```python
import matplotlib.pyplot as plt

def create_line_graph(x_values, y_values, title):
    plt.plot(x_values, y_values)
    plt.title(title)
    plt.xlabel("X-axis")
    plt.ylabel("Y-axis")
    plt.grid(True)
    plt.show()

def create_bar_graph(x_values, y_values, title):
    plt.bar(x_values, y_values)
    plt.title(title)
    plt.xlabel("X-axis")
    plt.ylabel("Y-axis")
    plt.show()

def create_scatter_plot(x_values, y_values, title):
    plt.scatter(x_values, y_values)
    plt.title(title)
    plt.xlabel("X-axis")
    plt.ylabel("Y-axis")
    plt.show()

def get_user_input():
    graph_type = input("Enter the type of graph (line/bar/scatter): ").lower()
    x_values = input("Enter x values separated by spaces: ").split()
    x_values = [float(x) for x in x_values]
    y_values = input("Enter y values separated by spaces: ").split()
    y_values = [float(y) for y in y_values]
    title = input("Enter graph title: ")
    return graph_type, x_values, y_values, title

def create_graph(graph_type, x_values, y_values, title):
    if graph_type == "line":
        create_line_graph(x_values, y_values, title)
    elif graph_type == "bar":
        create_bar_graph(x_values, y_values, title)
    elif graph_type == "scatter":
        create_scatter_plot(x_values, y_values, title)
    else:
        print("Invalid graph type. Please choose from line, bar, or scatter.")

# Get user input
graph_type, x_values, y_values, title = get_user_input()

# Create the graph
create_graph(graph_type, x_values, y_values, title)
```
